<h1 style = 'font-family: Times New Roman'> <b>|</b><i> About Grand Slam</i></h1>

<br>

> <h4 style = 'font-family: Times New Roman'>The Grand Slam tournaments, also referred to as majors, are the world's four most important annual professional tennis tournaments.<br><br> They offer the most ranking points, prize money, public and media attention, the greatest strength and size of field, and the longest matches for men.<br><br>
    The four Grand Slam tournaments are the <b>Australian Open</b> in January, the <b>French Open</b> from late May to early June, <b>Wimbledon</b> in late June to early July, <br><br>and the <b>US Open</b> in Augustâ€“September, with each played over two weeks. The Australian and the United States tournaments are played on <u>hard courts</u>,<br><br> the French on <u>clay</u>, and Wimbledon on <u>grass</u>. Wimbledon is the oldest tournament, founded in 1877, followed by the US in 1881, <br><br>the French in 1891, and the Australian in 1905, but it wasn't until 1925 that all four were held as officially sanctioned majors.

<br><br>
    
  
<h1 style = 'font-family: Times New Roman'> <b>|</b><i> About Dataset</i></h1>

<br> 
9 columns in the dataset, below is <u>description of features</u>:

*  (1) <b>YEAR</b>: the year when the tournament started
    
    
*  (2) <b>TOURNAMENT</b>: type of tournament(Wimbledon, u.s.open, French open, Australian open)
    
    
*  (3) <b>WINNER</b>: Winner's name
    
    
*  (4) <b>RUNNER UP</b>: Runner-up's name
    
    
*  (5) <b>WINNER ATP RANKING</b>: winner's atp ranking (>=2007)
    
    
*  (6) <b>RUNNER UP ATP RANKING</b>: runner up's atp ranking (>=2007)
    
    
*  (7) <b>WINNER LEFT OR RIGHT HANDED</b>: winner(right or left handed)
    
    
*  (8) <b>TOURNAMENT SURFACE</b>: surface of tournament 
    
    
*  (9) <b>WINNER PRIZE</b>: winner's prize of each tournament

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/kaggle/input/mens-tennis-grand-slam-winner-dataset/Mens_Tennis_Grand_Slam_Winner.csv')
df.head()
```

<a id="2.1"></a>
> <h2 style = 'font-family: Times New Roman'><b> 2.1. Winner Analysis </b></h2>

```python
WINNER_RUNNER_merged = pd.merge(df['RUNNER-UP'].value_counts().reset_index(), df['WINNER'].value_counts().reset_index(), on = 'index')
WINNER_RUNNER_merged['Total_Win'] = WINNER_RUNNER_merged['RUNNER-UP'] + WINNER_RUNNER_merged['WINNER']
WINNER_RUNNER_merged = WINNER_RUNNER_merged.sort_values('Total_Win', ascending = False)

win = WINNER_RUNNER_merged.iloc[:10]
win['Total_Win'] = win['Total_Win'].astype(int)
```

top 10 total win count

```python
colors = px.colors.qualitative.T10

fig = go.Figure(data = [
    go.Line(x = win['index'], y = win['RUNNER-UP'], line_color = colors[0], name = "RUNNER-UP"),
    go.Line(x = win['index'], y = win['WINNER'], line_color = colors[1],name="WINNER"),
    go.Line(x = win['index'], y = win['Total_Win'], line_color = colors[2], name='Total_Win')
])

fig.update_layout(
    xaxis_title="Player's Name",
    yaxis_title="Count of Wins",
    font = dict(size=17,family="Franklin Gothic"),
    template = 'ggplot2',
    title = 'Total / Winner / Runner-Up Wins')

fig.show()
```

```python
tournament = df.groupby(['TOURNAMENT','WINNER']).size().reset_index()
tournament = tournament[tournament['WINNER'].isin(win['index'])]

AO = tournament[tournament['TOURNAMENT'] == 'Australian Open'].sort_values(0, ascending = False)
FO = tournament[tournament['TOURNAMENT'] == 'French Open'].sort_values(0, ascending = False)
UO = tournament[tournament['TOURNAMENT'] == 'U.S. Open'].sort_values(0, ascending = False)
WB = tournament[tournament['TOURNAMENT'] == 'Wimbledon'].sort_values(0, ascending = False)

new = pd.concat([AO,FO,UO,WB])


new.columns = ['TOURNAMENT', 'WINNER', 'COUNT']

color = ['blue', 'brown', 'darkblue', 'green']
fig = px.scatter(new, x = 'WINNER', y = 'COUNT', size = 'COUNT', hover_name = 'TOURNAMENT', color = 'TOURNAMENT', 
          color_discrete_sequence=color, template = 'ggplot2')

fig.update_layout(
    xaxis_title="Winner's Name",
    yaxis_title="Wins Count",
    font = dict(size=17,family="Franklin Gothic"),
    template = 'ggplot2',
    title = 'Wins by Each Tournaments')

fig.show()
```

<a id="2.2"></a>
> <h1 style = 'font-family: Times New Roman'><b> 2.2. Prize Analysis </b></h1>

```python
a = df.groupby('WINNER')['WINNER_PRIZE'].sum().reset_index()
a['name'] = a['WINNER']
b = df.groupby('RUNNER-UP')['WINNER_PRIZE'].sum().reset_index()
b['name'] = b['RUNNER-UP']
prize_df = pd.merge(a, b, on ='name')
prize_df = prize_df[['WINNER', 'WINNER_PRIZE_x', 'WINNER_PRIZE_y']]
prize_df.columns = ['Player', 'TOTAL_WINNER_PRIZE', 'TOTAL_RUNNER_UP_PRIZE']

prize_df['TOTAL_PRIZE'] = prize_df['TOTAL_WINNER_PRIZE'] + prize_df['TOTAL_RUNNER_UP_PRIZE']
prize_df = prize_df.sort_values('TOTAL_PRIZE',ascending = False)

prize = prize_df.iloc[:10].reset_index().drop('index',axis=1)
prize.head()
```

```python
fig = go.Figure(data=[
    go.Bar(name='WINNER_PRIZE', 
           x=prize['Player'], y=prize['TOTAL_WINNER_PRIZE'],
           text=prize['TOTAL_WINNER_PRIZE'], 
           marker_color=colors[0]),
    
    go.Bar(name='RUNNER_UP_PRIZE', 
           x=prize['Player'], y=prize['TOTAL_RUNNER_UP_PRIZE'],
           text=prize['TOTAL_RUNNER_UP_PRIZE'], 
           marker_color=colors[1]),
    
    go.Bar(name='TOTAL_PRIZE', 
           x=prize['Player'], y=prize['TOTAL_PRIZE'],
           text=prize['TOTAL_PRIZE'], 
           marker_color=colors[5]),
])

fig.update_layout(xaxis_tickangle=-45, title='Winner & Runner-Up Prizes',
                  font = dict(size=17, family = 'Franklin Gothic'), template = "ggplot2")
fig.show()
```

```python
nationality = df['WINNER_NATIONALITY'].value_counts()[:8]
fig = px.pie(values=nationality.values, 
             names=nationality.index, 
             color_discrete_sequence=px.colors.qualitative.T10,
             title= "Distribution of Winner's Nationality",)
fig.update_traces(textinfo='label+percent+value', textfont_size=18,
                  
                  marker=dict(line=dict(color='#100000', width=0.2)))

fig.data[0].marker.line.width = 2
fig.data[0].marker.line.color='white'
fig.update_layout(font=dict(size=20,family="Franklin Gothic"))
fig.show()
```

```python
total = df.groupby(['YEAR'])['WINNER_PRIZE'].sum()
usopen = df[df['TOURNAMENT']=='U.S. Open'].groupby(['YEAR'])['WINNER_PRIZE'].sum()
frenchopen = df[df['TOURNAMENT']=='French Open'].groupby(['YEAR'])['WINNER_PRIZE'].sum()
wimbledon = df[df['TOURNAMENT']=='Wimbledon'].groupby(['YEAR'])['WINNER_PRIZE'].sum()
australianopen = df[df['TOURNAMENT']=='Australian Open'].groupby(['YEAR'])['WINNER_PRIZE'].sum()
```

```python
colors = px.colors.qualitative.T10

fig = go.Figure(data = [
#     go.Line(x = total.index, y = total.values, line_color = colors[0], name = "Total"),
    go.Line(x = usopen.index, y = usopen.values, line_color = colors[1], name = "U.S. Open"),
    go.Line(x = frenchopen.index, y = frenchopen.values, line_color = colors[2], name = "French Open"),
    go.Line(x = wimbledon.index, y = wimbledon.values, line_color = colors[3], name = "Wimbledon Open"),
    go.Line(x = australianopen.index, y = australianopen.values, line_color = colors[4], name = "Australian Open"),
])

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Prizes",
    font = dict(size=17,family="Franklin Gothic"),
    template = 'ggplot2',
    title = 'Distribution of Prizes of Each Tournaments')

fig.show()
```
